```python
train_df = pd.read_csv('../input/widsdatathon2023/train_data.csv')
test_df = pd.read_csv('../input/widsdatathon2023/test_data.csv')
submit = pd.read_csv('../input/widsdatathon2023/sample_solution.csv')
```

```python
train_df.startdate = pd.to_datetime(train_df.startdate)
test_df.startdate = pd.to_datetime(test_df.startdate)
import datetime
def remove_startdate(dataframe):
    dataframe['date'] = (dataframe['startdate'] - dataframe['startdate'].min()).dt.days
    dataframe = dataframe.drop('startdate', axis= 1).copy()
    return dataframe
train_df = remove_startdate(train_df)
test_df = remove_startdate(test_df)
```

```python
data = pd.concat([train_df, test_df],axis = 0)
data
```

```python
data['lat_lon'] = data.groupby(['lat', 'lon']).ngroup()
print(data.lat_lon.nunique())
```

```python
data['lat'] = round(data['lat'], 14)
data['lon'] = round(data['lon'], 14)
data['lat_lon'] = data.groupby(['lat', 'lon']).ngroup()
print(data.lat_lon.nunique())
```

```python
data = data.drop('lat', axis= 1)
data = data.drop('lon', axis= 1)

from sklearn.preprocessing import LabelEncoder  
le = LabelEncoder()

data['climateregions__climateregion'] = le.fit_transform(data['climateregions__climateregion'])
print(data['climateregions__climateregion'])

train_df = data[:len(train_df)]
test_df = data[len(train_df):]
```

```python
def null_cols(dataframe):
    null_cols = dataframe.columns[dataframe.isnull().any()]
    return null_cols

null_columns_train = null_cols(train_df)
null_columns_train
```

```python
null_cols(test_df)
```

```python
def fill_na(dataframe, null_columns):
    nulls = null_columns
    for col in nulls:
            dataframe[col].fillna(dataframe[col].mean(), inplace=True)
    return dataframe

train_df = fill_na(train_df, null_columns_train)
null_cols(train_df)
```

```python
y_train = train_df['contest-tmp2m-14d__tmp2m']
X_train = train_df.drop('contest-tmp2m-14d__tmp2m', axis= 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train.shape
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
test_df = test_df.drop('contest-tmp2m-14d__tmp2m', axis= 1)
```

```python
pred = model.predict(test_df)
```

```python
pred = pd.DataFrame(pred, columns=['contest-tmp2m-14d__tmp2m'])
pred['index'] = submit['index']
```

```python
pred.to_csv("submission.csv", index= False)
```
